# Bar Chart:

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["Python","ML","AI","DS","DL"]
class_1_student=[15,10,40,25,10]
class_2_Student=[30,20,5,25,20]
class_3_student=[5,20,30,25,20]
```

```python
plt.bar(classes,class_1_student,color='m')
plt.show()
```

```python
plt.bar(classes,class_2_Student,color='y')
plt.show()
```

```python
plt.bar(classes,class_3_student,color='g')
plt.show()
```

```python
plt.barh(classes,class_1_student)
plt.show()
```

```python
#width parameter:
plt.bar(classes,class_1_student,width=0.2)
plt.show()
```

```python
#align parameter:
plt.bar(classes,class_1_student,width=0.2,align='edge')
plt.show()
```

```python
#edgecolor parameter:
plt.bar(classes,class_1_student,width=0.2,align='edge',color='y',edgecolor='b')
plt.show()
```

```python
#linewidth parameter:
plt.figure(figsize=(10,8))
plt.bar(classes,class_1_student,width=0.2,align='edge',color='y',edgecolor='b',linewidth=4)
plt.show()
```

```python
#alpha parameter:0.4
plt.figure(figsize=(10,8))
plt.bar(classes,class_1_student,width=0.2,align='edge',color='y',edgecolor='b',linewidth=4,alpha=0.4)
plt.show()
```

```python
#alpha parameter:0.8
plt.figure(figsize=(10,8))
plt.bar(classes,class_1_student,width=0.2,align='edge',color='y',edgecolor='b',linewidth=4,alpha=0.8)
plt.show()
```

```python
#comparing all 3 class students,adding xticks
plt.figure(figsize=(15,10))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class_1_student,width,color='g')
plt.bar(class_index+width,class_2_Student,width,color='y')
plt.bar(class_index+width+width,class_3_student,width,color='b')
plt.xticks(class_index+width,classes)      #replace 0,1,2,3,4 with the class name i.e python,ai,ml,ds,dl
plt.show()
```

```python
#adding title,xlabel,ylabel and legend:
plt.figure(figsize=(15,10))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class_1_student,width,color='g')
plt.bar(class_index+width,class_2_Student,width,color='y')
plt.bar(class_index+width+width,class_3_student,width,color='b')
plt.xticks(class_index+width,classes)
plt.title("Number of Students in different classes",fontsize=20)
plt.xlabel("classes",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["1class student","2class student","3class student"],loc=1)
plt.show()
```

```python
#Horizontal bar chart::
plt.figure(figsize=(15,10))
class_index=np.arange(len(classes))
width=0.2
plt.barh(class_index,class_1_student,width,color='g')
plt.barh(class_index+width,class_2_Student,width,color='y')
plt.barh(class_index+width+width,class_3_student,width,color='m')
plt.yticks(class_index+width,classes)
plt.title("Number of Students in different classes",fontsize=25)
plt.ylabel("classes",fontsize=20)
plt.xlabel("no of students",fontsize=20)
plt.legend(["1class student","2class student","3class student"],loc=1)
plt.show()
```
