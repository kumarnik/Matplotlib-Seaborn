```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/nikhil/Desktop/google-play-store-apps/googleplaystore.csv",nrows=1000)
```

```python
df.head()
```

```python
df.shape
```

```python
x=df['Rating']
y=df['Reviews']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.title("rating vs reviews")
plt.xlabel("rating")
plt.ylabel("reviews")
plt.show()
```

```python
#marker added:*,color added green
plt.figure(figsize=(10,8))
plt.scatter(x,y,color='g',marker="*")
plt.title("rating vs reviews")
plt.xlabel("rating")
plt.ylabel("reviews")
plt.show()
```

```python
#s:for the marker size,fontsize added
plt.figure(figsize=(10,8))
plt.scatter(x,y,color='g',marker="*",s=100)
plt.title("rating vs reviews",fontsize=20)
plt.xlabel("rating",fontsize=15)
plt.ylabel("reviews",fontsize=15)
plt.show()
```

```python
#linewidth paramter added:
plt.figure(figsize=(20,10))
plt.scatter(x,y,color='g',marker="*",s=100,linewidth=5)
plt.title("rating vs reviews",fontsize=20)
plt.xlabel("rating",fontsize=15)
plt.ylabel("reviews",fontsize=15)
plt.show()
```

```python
#alpha parmeter added:
plt.figure(figsize=(20,8))
plt.scatter(x,y,color='g',marker="*",s=100,alpha=0.4)
plt.title("rating vs reviews",fontsize=20)
plt.xlabel("rating",fontsize=15)
plt.ylabel("reviews",fontsize=15)
plt.show()
```

```python
#edgecolor added:
plt.figure(figsize=(20,8))
plt.scatter(x,y,color='g',marker="*",s=100,alpha=0.4,linewidth=10,edgecolors='y')
plt.title("rating vs reviews",fontsize=20)
plt.xlabel("rating",fontsize=15)
plt.ylabel("reviews",fontsize=15)
plt.show()
```

```python
#adding 2 scatter plots:
plt.figure(figsize=(20,8))
plt.scatter(x,y,color='r',marker="*",s=100,alpha=0.5,linewidth=10,edgecolors='g')
plt.scatter(x,df['Installs'],color='y',marker="o",s=100,alpha=0.5,linewidth=10,edgecolors='c')
plt.title("rating vs reviews &installs",fontsize=20)
plt.xlabel("rating",fontsize=15)
plt.ylabel("reviews and installs",fontsize=15)
plt.show()
```
