## Subplot:

```python
import matplotlib.pyplot as plt
```

```python
#2 rows and 2 columns and then the index value:1,2,3,4
plt.figure(figsize=(20,8))
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.show()
```

```python
#2 rows and 3 columns:
plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.subplot(2,3,2)
plt.subplot(2,3,3)
plt.subplot(2,3,4)
plt.subplot(2,3,5)
plt.subplot(2,3,6)
plt.show()
```

```python
#example:
plt.figure(figsize=(20,8))
plt.subplot(2,2,1)
plt.pie([10,20,15],labels=["car","bikes","trucks"],radius=1.4)
plt.subplot(2,2,2)
plt.plot([1,2,3,4],[10,20,5,30],color="r")
plt.subplot(2,2,3)
plt.bar(["IT","automobile","film"],[60,30,50],color="m")
plt.subplot(2,2,4)
plt.hist([1,2,3,1,1,3,4,1,4],color="y")
plt.show()
```
