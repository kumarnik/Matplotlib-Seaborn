## Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
classes=["Python","ML","AI","DS","DL"]
no_of_student=[25,20,40,10,5]
```

```python
plt.figure(figsize=(7,7))
plt.pie(no_of_student)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(no_of_student,labels=classes)
plt.show()
```

```python
#explode parameter :to enhance any slice
plt.figure(figsize=(7,7))
explode=[0,0,0,0,0]
plt.pie(no_of_student,labels=classes,explode=explode)
plt.show()     #no change as explode value is 0 for all 
```

```python
#explode parameter :to enhance any slice
plt.figure(figsize=(7,7))
explode=[0,0,0.2,0,0]
plt.pie(no_of_student,labels=classes,explode=explode)
plt.show() 
```

```python
#adding colour attribute:
plt.figure(figsize=(7,7))
colors=['c','b','r','y','g']
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors)
plt.show()
```

```python
#adding auto percentage attribute:
plt.figure(figsize=(7,7))
colors=['c','b','r','y','g']
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%")
plt.show()
```

```python
#added shadow attribute:
plt.figure(figsize=(7,7))
colors=['c','b','r','y','g']
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
#added radius attribute:to inc/dec radius 
colors=['c','b','r','y','g']
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=2.5)
plt.show()
```

```python
#startangle attribute:start the label with that angle
colors=['c','b','r','y','g']
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=1.5,startangle=80)
plt.show()
```

```python
#textprops attribute:text properties
colors=['c','b','m','y','g']
textprops={"fontsize":12}
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=2.5,startangle=80,textprops=textprops)
plt.show()
```

```python
#pct distance attribute: percentage how much far from the center
colors=['c','b','m','y','g']
textprops={"fontsize":12}
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=2.5,startangle=80,textprops=textprops,pctdistance=0.8)
plt.show()
```

```python
#label distance attribute: label how much far from the center
colors=['c','b','m','y','g']
textprops={"fontsize":12}
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=2.5,startangle=80,textprops=textprops,pctdistance=0.8,
       labeldistance=1.4)
plt.show()
```

```python
#label should not be too far nor too close to the center:
colors=['c','b','m','y','g']
textprops={"fontsize":12}
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=2.5,startangle=80,textprops=textprops,pctdistance=0.8,
       labeldistance=1.1)
plt.show()
```

```python
#wedge properties:
colors=['c','b','m','y','g']
wedgeprops={"linewidth":8,"width":1,"edgecolor":"orange"}
textprops={"fontsize":12}
explode=[0,0,0.1,0,0]
plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=2.5,startangle=80,textprops=textprops,pctdistance=0.8,
       labeldistance=1.1,wedgeprops=wedgeprops)
plt.show()
```

```python
#adding legend:bbox_to anchor to shift legend out of the pie:
colors=['c','b','m','y','g']
textprops={"fontsize":12}
explode=[0,0,0.1,0,0]
p=plt.pie(no_of_student,labels=classes,explode=explode,colors=colors,
       autopct="%0.2f%%",shadow=True,radius=2.5,startangle=80,textprops=textprops,pctdistance=0.8,
       labeldistance=1.1)
plt.legend(bbox_to_anchor=(2,1.5),fontsize=15, loc="center right")
plt.show()
```
