# Histogram:

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#random module to genrate random age for ml and python 100 student:
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
```

```python
print(py_students_age)
```

```python
plt.hist(ml_students_age)
plt.show()
```

```python
#histogram for python students age:
plt.hist(py_students_age)
plt.show()
```

```python
#increasing figure size
#adding title,xlabel,ylabel and legend

plt.figure(figsize=(8,6))
plt.hist(ml_students_age)
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student"])
plt.show()
```

```python
#adding bins:
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(8,6))
plt.hist(ml_students_age,bins)
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student"])
plt.show()
```

```python
#adding rwidth,color
plt.figure(figsize=(8,6))
plt.hist(ml_students_age,bins,rwidth=0.9,color='m')
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student"])
plt.show()
```

```python
#adding orientation
plt.figure(figsize=(8,6))
plt.hist(ml_students_age,bins,rwidth=0.9,color='m',orientation="horizontal")
plt.title("ML Students Age",fontsize=20)
plt.ylabel("age",fontsize=15)
plt.xlabel("no of students",fontsize=15)
plt.legend(["ml-student"])
plt.show()
```

```python
#adding histtype:
plt.figure(figsize=(8,6))
plt.hist(ml_students_age,bins,rwidth=0.9,color='m',histtype='step')
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student"])
plt.show()
```

```python
#adding histtype=barstacked:
plt.figure(figsize=(8,6))
plt.hist(ml_students_age,bins,rwidth=0.9,color='m',histtype='barstacked')
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student"])
plt.show()
```

```python
#adding histtype=stepfilled:
plt.figure(figsize=(8,6))
plt.hist(ml_students_age,bins,rwidth=0.9,color='m',histtype='stepfilled')
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student"])
plt.show()
```

```python
#now adding ml and py students age into single histogram:
plt.figure(figsize=(8,6))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=['m','b'])
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student","py-student"],loc=4)

plt.show()
```

```python
#style module:
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(8,6))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=['m','b'])
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student","py-student"],loc=4)
plt.show()
```

```python
#adding grid function and change the legend loc to 0 i.e upper left:
plt.figure(figsize=(8,6))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=['m','b'])
plt.title("ML Students Age",fontsize=20)
plt.xlabel("age",fontsize=15)
plt.ylabel("no of students",fontsize=15)
plt.legend(["ml-student","py-student"],loc=0)
plt.grid(color='y', linestyle='--', linewidth=1)
plt.show()
```
