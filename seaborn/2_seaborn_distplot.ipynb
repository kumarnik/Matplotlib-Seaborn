## Dist plot:

```python
import seaborn as sns
from scipy.stats import norm 
import matplotlib.pyplot as plt
```

```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df.head()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(tips_df["size"])
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(tips_df["size"],bins=10)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(tips_df["size"],bins=3)
plt.show()
```

```python
# now visualize total_bill:

plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"])

plt.show()
```

```python
#20 bins:
plt.figure(figsize=(12,8))
sns.distplot(tips_df["total_bill"],
             bins=20)
plt.show()
```

```python
#50 bins:
plt.figure(figsize=(12,8))
sns.distplot(tips_df["total_bill"],
             bins=50)
plt.show()
```

```python
#
plt.figure(figsize=(12,8))
sns.distplot(tips_df["total_bill"],
             bins=[30,40,50])
plt.show()
```

```python
"""
Signature:
sns.distplot(
    a,
    bins=None,
    hist=True,
    kde=True,
    rug=False,
    fit=None,
    hist_kws=None,
    kde_kws=None,
    rug_kws=None,
    fit_kws=None,
    color=None,
    vertical=False,
    norm_hist=False,
    axlabel=None,
    label=None,
    ax=None,
)

"""
```

```python
#hist:

plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"],bins=[30,40,50],
             hist=False)

plt.show()
```

```python
#kde:kernel density estimator

plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"],bins=[30,40,50],
             kde=False)
plt.show()
```

```python
#rug:


plt.figure(figsize=(12,8))
sns.distplot(tips_df["total_bill"],bins=[30,40,50],
             rug=True)
plt.show()
```

```python
#fit=norm:

plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"],bins=[30,40,50],
             fit=norm)
plt.show()
```

```python
#only norm and not kde:

plt.figure(figsize=(12,8))


sns.distplot(tips_df["total_bill"], bins=[30,40,50],
             fit=norm,
             kde=False)

plt.show()
```

```python
#color:

plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"],bins=[30,40,50],
             color="r")
plt.show()
```

```python
# vertical attribute:
plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"],bins=[30,40,50],color='g',
             vertical=True)
plt.show()
```

```python
#axlabel:

plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"],bins=[30,40,50],color="r",
             axlabel="Total-BILL")

plt.show()
```

```python
#label:
plt.figure(figsize=(12,8))

sns.distplot(tips_df["total_bill"],bins=[30,40,50],color="m",
             label="total-bill")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],label="total_bill",color='g')
plt.title("Histogram of Total_bill")
plt.legend()
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins)
plt.title("Histogram of Total_bill")
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins)
plt.title("Histogram of Total_bill")
plt.xticks(bins)
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins,
            hist_kws={"color":'#DC143C' ,"edgecolor":'#aaff00',"linewidth":5,"linestyle":"--"})
plt.title("Histogram of Total_bill")
plt.xticks(bins)
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins,
            hist_kws={"color":'#DC143C' ,"edgecolor":'#aaff00',"linewidth":5,"linestyle":"--","alpha":0.7})
plt.title("Histogram of Total_bill")
plt.xticks(bins)
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins,
            hist_kws={"color":'#DC143C' ,"edgecolor":'#aaff00',"linewidth":5,"linestyle":"--","alpha":0.8},
            kde_kws={"color":'#8e00ce' ,"linewidth":2,"linestyle":"--","alpha":0.9})
plt.title("Histogram of Total_bill")
plt.xticks(bins)
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins,
            hist_kws={"color":'#DC143C' ,"edgecolor":'#aaff00',"linewidth":5,"linestyle":"--","alpha":0.8},
            kde_kws={"color":'#8e00ce' ,"linewidth":2,"linestyle":"--","alpha":0.9},
             rug=True,
            rug_kws={"color":'#0426d0' ,"edgecolor":'#00dbff',"linewidth":2,"linestyle":"--","alpha":0.8})
plt.title("Histogram of Total_bill")
plt.xticks(bins)
plt.show()
```

```python
# fit attribute added,
# and when we add fit we just do kde =false

bins=[1,5,10,15,20,25,30,35,40,45,50,55,60,65]
plt.figure(figsize=(12,8))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins,
            hist_kws={"color":'#DC143C' ,"edgecolor":'#aaff00',"linewidth":5,"linestyle":"--","alpha":0.8},
             kde=False,
             fit=norm,
             fit_kws={"color":'#8e00ce' ,"linewidth":2,"linestyle":"--","alpha":0.9},
            #kde_kws={"color":'#8e00ce' ,"linewidth":2,"linestyle":"--","alpha":0.9},
             rug=True,
            rug_kws={"color":'#0426d0' ,"edgecolor":'#00dbff',"linewidth":2,"linestyle":"--","alpha":0.8})
plt.title("Histogram of Total_bill")
plt.xticks(bins)
plt.show()
```
